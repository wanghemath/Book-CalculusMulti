## 1. Functions with one input and many inputs

Single-variable calculus studies functions such as

$$
f:\mathbb R\to \mathbb R, \qquad x\mapsto f(x).
$$

Multivariable calculus studies functions such as

$$
f:\mathbb R^2\to \mathbb R, \qquad (x,y)\mapsto f(x,y),
$$

and more generally

$$
f:\mathbb R^n\to \mathbb R, \qquad x=(x_1,\dots,x_n)\mapsto f(x).
$$

A key shift is this: one-dimensional inputs move left or right, but multivariable inputs can move in many directions.


```python
def f_one_variable(x):
    '''A one-variable function f: R -> R.'''
    return x**3 - 3*x

x = np.linspace(-3, 3, 500)
y = f_one_variable(x)

plt.figure(figsize=(7, 4))
plt.plot(x, y)
plt.axhline(0, linewidth=0.8)
plt.axvline(0, linewidth=0.8)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("A one-variable function: f(x)=x^3-3x")
plt.grid(True, alpha=0.3)
plt.show()
```

## 2. Vectors, points, and distances in $\mathbb R^n$

The same tuple can have different meanings depending on context. For example,

$$
(2,-1,3)
$$

could be a point in physical space, a displacement vector, three measurements, or three model parameters.

The Euclidean norm of $v=(v_1,\dots,v_n)$ is

$$
\|v\|=\sqrt{v_1^2+\cdots+v_n^2}.
$$

The Euclidean distance between points $a$ and $b$ is

$$
\operatorname{dist}(a,b)=\|b-a\|.
$$


```python
def norm(v):
    '''Euclidean norm of a vector.'''
    v = np.asarray(v, dtype=float)
    return np.sqrt(np.sum(v**2))


def distance(a, b):
    '''Euclidean distance between two points.'''
    a = np.asarray(a, dtype=float)
    b = np.asarray(b, dtype=float)
    return norm(b - a)

v = np.array([2, -1, 3])
a = np.array([1, 1, 0])
b = np.array([3, 0, 3])

print("v =", v)
print("||v|| =", norm(v))
print("a =", a)
print("b =", b)
print("b - a =", b - a)
print("dist(a,b) =", distance(a, b))
```

## 3. Four representations of a two-variable function

Consider the temperature model

$$
T(x,y)=80-2x^2-y^2.
$$

This function can be represented verbally, algebraically, numerically, visually, and computationally.

- **Verbal:** temperature is largest near the origin and decreases quadratically away from the origin.
- **Algebraic:** $T(x,y)=80-2x^2-y^2$.
- **Numerical:** values can be sampled on a grid.
- **Visual:** a surface plot, contour plot, or heatmap.
- **Computational:** a Python function.


```python
def T(x, y):
    return 80 - 2*x**2 - y**2

# Numerical table on a small grid.
x_values = np.array([-2, -1, 0, 1, 2])
y_values = np.array([-2, -1, 0, 1, 2])

print("Table of T(x,y) values")
print("Rows: y values; columns: x values")
print("x values:", x_values)
for y0 in y_values:
    row = [T(x0, y0) for x0 in x_values]
    print(f"y = {y0:>2}: {row}")
```

```python
x = np.linspace(-5, 5, 250)
y = np.linspace(-5, 5, 250)
X, Y = np.meshgrid(x, y)
Z = T(X, Y)

plt.figure(figsize=(6, 5))
cs = plt.contour(X, Y, Z, levels=16)
plt.clabel(cs, inline=True, fontsize=8)
plt.xlabel("x")
plt.ylabel("y")
plt.gca().set_aspect("equal")
plt.title("Contour plot of T(x,y)=80-2x^2-y^2")
plt.grid(True, alpha=0.3)
plt.show()
```

## 4. Surfaces and level curves

For $f:\mathbb R^2\to \mathbb R$, the graph

$$
z=f(x,y)
$$

is a surface in $\mathbb R^3$. A level curve is the curve

$$
f(x,y)=c.
$$

The next helper function shows both the surface and the level curves.


```python
def plot_surface_and_contours(f, xlim=(-3, 3), ylim=(-3, 3), title="f(x,y)", levels=20):
    x = np.linspace(xlim[0], xlim[1], 180)
    y = np.linspace(ylim[0], ylim[1], 180)
    X, Y = np.meshgrid(x, y)
    Z = f(X, Y)

    fig = plt.figure(figsize=(7, 5))
    ax = fig.add_subplot(111, projection="3d")
    ax.plot_surface(X, Y, Z, linewidth=0, antialiased=True, alpha=0.9)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("z")
    ax.set_title("Surface: " + title)
    plt.show()

    plt.figure(figsize=(6, 5))
    cs = plt.contour(X, Y, Z, levels=levels)
    plt.clabel(cs, inline=True, fontsize=8)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.gca().set_aspect("equal")
    plt.title("Level curves: " + title)
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
plot_surface_and_contours(lambda x, y: x**2 + y**2, title="z = x^2 + y^2", levels=12)
```

```python
plot_surface_and_contours(lambda x, y: x**2 - y**2, title="z = x^2 - y^2", levels=21)
```

## 5. Partial derivatives and numerical gradients

For a function $f(x,y)$, the partial derivative with respect to $x$ measures change in the $x$ direction while $y$ is held fixed:

$$
\frac{\partial f}{\partial x}(a,b)
\approx
\frac{f(a+h,b)-f(a-h,b)}{2h}.
$$

Similarly,

$$
\frac{\partial f}{\partial y}(a,b)
\approx
\frac{f(a,b+h)-f(a,b-h)}{2h}.
$$

The gradient is

$$
\nabla f(a,b)=\left\langle f_x(a,b), f_y(a,b)\right\rangle.
$$


```python
def numerical_gradient_2d(f, x0, y0, h=1e-5):
    '''Centered-difference approximation to the gradient of f(x,y).'''
    fx = (f(x0 + h, y0) - f(x0 - h, y0)) / (2*h)
    fy = (f(x0, y0 + h) - f(x0, y0 - h)) / (2*h)
    return np.array([fx, fy])


def f_quadratic(x, y):
    return x**2 + 2*y**2

point = (1.0, 1.0)
print("Numerical gradient at (1,1):", numerical_gradient_2d(f_quadratic, *point))
print("Exact gradient at (1,1):    ", np.array([2*point[0], 4*point[1]]))
```

## 7. Integrals as accumulation over regions

Single-variable integration accumulates over intervals. Double integration accumulates over regions.

For a nonnegative function $f(x,y)$ over a rectangle $R=[a,b]\times[c,d]$, the double integral

$$
\iint_R f(x,y)\,dA
$$

can be approximated by a Riemann sum:

$$
\iint_R f(x,y)\,dA\approx \sum_{i,j} f(x_i^*,y_j^*)\Delta A.
$$

If $f(x,y)\ge 0$, this approximates the volume under $z=f(x,y)$ and above $R$.


```python
def midpoint_riemann_sum_2d(f, xlim, ylim, nx=20, ny=20):
    '''Midpoint-rule approximation of a double integral over a rectangle.'''
    a, b = xlim
    c, d = ylim
    dx = (b - a) / nx
    dy = (d - c) / ny
    x_mid = a + (np.arange(nx) + 0.5) * dx
    y_mid = c + (np.arange(ny) + 0.5) * dy
    Xmid, Ymid = np.meshgrid(x_mid, y_mid)
    return np.sum(f(Xmid, Ymid)) * dx * dy


def f_volume(x, y):
    return 16 - x**2 - 2*y**2

for n in [4, 8, 16, 32, 64]:
    approx = midpoint_riemann_sum_2d(f_volume, (0, 2), (0, 2), nx=n, ny=n)
    print(f"n = {n:>2}, midpoint approximation = {approx:.6f}")
```

## 8. Loss functions as landscapes

A central modern application of multivariable calculus is optimization. Suppose we have data points $(x_i,y_i)$ and fit a line

$$
\widehat y = wx+b.
$$

The mean squared error loss is

$$
L(w,b)=\frac{1}{m}\sum_{i=1}^m (wx_i+b-y_i)^2.
$$

This is a scalar-valued function of two variables, $w$ and $b$. The point $(w,b)$ lives in parameter space.


```python
rng = np.random.default_rng(2321)

x_data = np.linspace(-2, 2, 25)
y_data = 1.5*x_data - 0.4 + rng.normal(scale=0.35, size=x_data.shape)

plt.figure(figsize=(6, 4))
plt.scatter(x_data, y_data)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Synthetic data for line fitting")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
def mse_loss(w, b, x=x_data, y=y_data):
    return np.mean((w*x + b - y)**2)

w_grid = np.linspace(-1, 4, 180)
b_grid = np.linspace(-2, 2, 180)
W, B = np.meshgrid(w_grid, b_grid)
Loss = np.zeros_like(W)

for i in range(W.shape[0]):
    for j in range(W.shape[1]):
        Loss[i, j] = mse_loss(W[i, j], B[i, j])

fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection="3d")
ax.plot_surface(W, B, Loss, linewidth=0, antialiased=True, alpha=0.9)
ax.set_xlabel("w")
ax.set_ylabel("b")
ax.set_zlabel("L(w,b)")
ax.set_title("Least-squares loss surface")
plt.show()

plt.figure(figsize=(6, 5))
cs = plt.contour(W, B, Loss, levels=25)
plt.clabel(cs, inline=True, fontsize=8)
plt.xlabel("w")
plt.ylabel("b")
plt.title("Contour plot of L(w,b)")
plt.grid(True, alpha=0.3)
plt.show()
```

## 9. A first gradient descent experiment

Gradient descent updates parameters by moving opposite the gradient:

$$
(w,b)\leftarrow (w,b)-\eta\nabla L(w,b),
$$

where $\eta>0$ is the learning rate.

For the mean squared error,

$$
L(w,b)=\frac{1}{m}\sum_{i=1}^m (wx_i+b-y_i)^2,
$$

we have

$$
\frac{\partial L}{\partial w}=\frac{2}{m}\sum_{i=1}^m (wx_i+b-y_i)x_i,
$$

and

$$
\frac{\partial L}{\partial b}=\frac{2}{m}\sum_{i=1}^m (wx_i+b-y_i).
$$


```python
def mse_gradient(w, b, x=x_data, y=y_data):
    errors = w*x + b - y
    dL_dw = 2*np.mean(errors*x)
    dL_db = 2*np.mean(errors)
    return np.array([dL_dw, dL_db])


def gradient_descent(start, eta=0.1, steps=40):
    point = np.array(start, dtype=float)
    path = [point.copy()]
    for _ in range(steps):
        grad = mse_gradient(point[0], point[1])
        point = point - eta*grad
        path.append(point.copy())
    return np.array(path)

path = gradient_descent(start=(-0.5, 1.5), eta=0.12, steps=35)

print("Start:", path[0])
print("End:", path[-1])
print("Least-squares solution:", np.array([w_hat, b_hat]))
print("Final loss:", mse_loss(path[-1, 0], path[-1, 1]))
```

```python
plt.figure(figsize=(6, 5))
cs = plt.contour(W, B, Loss, levels=25)
plt.plot(path[:, 0], path[:, 1], marker="o")
plt.scatter([w_hat], [b_hat], marker="x", s=120)
plt.xlabel("w")
plt.ylabel("b")
plt.title("Gradient descent path on the loss contours")
plt.grid(True, alpha=0.3)
plt.show()
```

### Task 9.1

Try learning rates $\eta=0.02$, $0.12$, and $0.45$ from the same starting point. Compare the paths and final losses.


```python
# Student workspace / reference solution for Task 9.1
etas = [0.02, 0.12, 0.45]
start = (-0.5, 1.5)
paths = []

plt.figure(figsize=(6, 5))
plt.contour(W, B, Loss, levels=25)

for eta in etas:
    p = gradient_descent(start=start, eta=eta, steps=35)
    paths.append(p)
    plt.plot(p[:, 0], p[:, 1], marker="o", label=f"eta={eta}")
    final_loss = mse_loss(p[-1, 0], p[-1, 1])
    print(f"eta={eta}: final point={p[-1]}, final loss={final_loss:.6f}")

plt.scatter([w_hat], [b_hat], marker="x", s=120, label="least-squares")
plt.xlabel("w")
plt.ylabel("b")
plt.title("Effect of learning rate")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

**Reference interpretation.**

A small learning rate moves slowly but stably. A moderate learning rate often reaches the minimizer faster. A learning rate that is too large may zigzag or fail to settle down. This is one reason optimization is both a calculus problem and a numerical computation problem.


## 10. High-dimensional thinking

Most pictures in multivariable calculus use $\mathbb R^2$ and $\mathbb R^3$, but modern applications often use $\mathbb R^n$ with large $n$.

A data vector may be

$$
x=(x_1,x_2,\dots,x_n).
$$

A linear prediction model can be written compactly as

$$
\widehat y = x\cdot \beta,
$$

where $\beta$ is a parameter vector. The loss function is a function of the parameter vector.


```python
# A small high-dimensional linear regression example.
rng = np.random.default_rng(2026)

m = 80      # number of observations
n = 6       # number of features
X = rng.normal(size=(m, n))
true_beta = np.array([1.5, -2.0, 0.0, 0.7, 0.0, 1.0])
y = X @ true_beta + rng.normal(scale=0.4, size=m)

def high_dim_loss(beta):
    errors = X @ beta - y
    return np.mean(errors**2)

def high_dim_grad(beta):
    errors = X @ beta - y
    return 2 * X.T @ errors / len(y)

beta = np.zeros(n)
eta = 0.08
loss_history = []

for _ in range(120):
    loss_history.append(high_dim_loss(beta))
    beta = beta - eta * high_dim_grad(beta)

print("True beta:     ", true_beta)
print("Estimated beta:", beta)
print("Final loss:    ", high_dim_loss(beta))

plt.figure(figsize=(6, 4))
plt.plot(loss_history)
plt.xlabel("iteration")
plt.ylabel("loss")
plt.title("Gradient descent in a 6-dimensional parameter space")
plt.grid(True, alpha=0.3)
plt.show()
```

### Task 10.1

Explain why the plot above shows a one-dimensional curve even though the optimization happens in a six-dimensional parameter space.


**Reference solution.**

The parameter vector $\beta$ moves in $\mathbb R^6$, so the path is six-dimensional. The plot shows only the scalar loss value at each iteration. In other words, the horizontal axis is iteration number, and the vertical axis is $L(\beta)$, not a coordinate of $\beta$.

```python
# Reference solution for Challenge 1
functions = [
    (lambda x, y: x**2 + 4*y**2, "x^2 + 4y^2"),
    (lambda x, y: x**2 - y**2, "x^2 - y^2"),
    (lambda x, y: np.exp(-x**2 - y**2), "exp(-x^2-y^2)"),
]

x = np.linspace(-3, 3, 250)
y = np.linspace(-3, 3, 250)
X, Y = np.meshgrid(x, y)

for func, title in functions:
    Z = func(X, Y)
    plt.figure(figsize=(6, 5))
    cs = plt.contour(X, Y, Z, levels=15)
    plt.clabel(cs, inline=True, fontsize=8)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.gca().set_aspect("equal")
    plt.title("Level curves of " + title)
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
# Reference solution for Challenge 2
def f_challenge(x, y):
    return x**2 * y + np.sin(x*y)

numeric_grad = numerical_gradient_2d(f_challenge, 1, 2)
exact_grad = np.array([
    2*1*2 + 2*np.cos(1*2),  # f_x = 2xy + y cos(xy)
    1**2 + 1*np.cos(1*2),   # f_y = x^2 + x cos(xy)
])

print("Numerical gradient:", numeric_grad)
print("Exact gradient:    ", exact_grad)
print("Difference:        ", numeric_grad - exact_grad)
```

```python
# Reference solution for Challenge 3
rng = np.random.default_rng(123)
x_ch = np.linspace(-2, 2, 30)
y_ch = 2*x_ch + 1 + rng.normal(scale=0.25, size=x_ch.shape)

A_ch = np.column_stack([x_ch, np.ones_like(x_ch)])
w_ch, b_ch = np.linalg.lstsq(A_ch, y_ch, rcond=None)[0]

print("Least-squares slope:", w_ch)
print("Least-squares intercept:", b_ch)

plt.figure(figsize=(6, 4))
plt.scatter(x_ch, y_ch)
plt.plot(x_ch, w_ch*x_ch + b_ch)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Challenge 3 data and fitted line")
plt.grid(True, alpha=0.3)
plt.show()
```
